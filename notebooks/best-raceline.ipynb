```python
import numpy as np
import matplotlib.pyplot as plt
from scripts.calculate_raceline import print_border, \
    create_track_polygon, calculate_best_raceline, show_racelines_diffs;
```

```python
track_name = 'A to Z - Speedway'
waypoints = np.load("../data/reInvent2019_wide.npy") # Passing the track waypoints

# Convert waypoints to Shapely objects
center_line = waypoints[:, 0:2]
inner_border = waypoints[:, 2:4]
outer_border = waypoints[:, 4:6]

road_poly = create_track_polygon(inner_border, outer_border)
road_poly
```

```python
# Calculating the best raceline
race_line = calculate_best_raceline(center_line, inner_border, outer_border)
```

```python
loop_race_line = np.append(race_line, [race_line[0]], axis=0)

show_racelines_diffs(center_line, loop_race_line)

fig = plt.figure(1, figsize=(16, 10))
ax = fig.add_subplot(111, facecolor='white')
plt.axis('equal')
print_border(ax, loop_race_line, inner_border, outer_border)
```
